```python
import pandas as pd

import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/Users/larsheijnen/VSCode_S2/Thesis/data/Age_at_visit_21Mar2025.csv'
data = pd.read_csv(file_path)

# Plot the age distribution
plt.figure(figsize=(10, 6))
plt.hist(data['AGE_AT_VISIT'], bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age at Visit')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Generate summary statistics for the AGE_AT_VISIT column, including the median
age_statistics = data['AGE_AT_VISIT'].describe()
median_age = data['AGE_AT_VISIT'].median()
print(age_statistics)
print(f"Median: {median_age}")
```
